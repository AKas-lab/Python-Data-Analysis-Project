```python
# Task 7: Geographic Analysis of Python Developers’ Compensation
```

```python
# filter respondents who work with Python
worked_with_python = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
# select the required columns
py_df = df.loc[worked_with_python, ['Country', 'ConvertedCompYearly']]
```

```python
# remove rows without salary information
py_df = py_df.dropna(subset=['ConvertedCompYearly'])
```

```python
# group by country and calculate the mean and median
py_progress_py_country = (
    py_df
    .groupby('Country')['ConvertedCompYearly']
    .agg(['mean', 'median'])
    .reset_index()
)
```

```python
# Task 9: Analysis of Python Popularity by Age Groups
```

```python
# filter respondents who work with Python
worked_with_python = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
```

```python
# keep the required columns
age_df = df[['Age', 'LanguageHaveWorkedWith']].copy()
age_df['WorkedWithPython'] = worked_with_python
```

```python
# remove respondents without age information
age_df = age_df.dropna(subset=['Age'])
```

```python
# group and count
age_stats = (
    age_df
    .groupby('Age')
    .agg(
        total_respondents=('WorkedWithPython', 'count'),
        python_users=('WorkedWithPython', 'sum')
    )
)
```

```python
# calculate the percentage
age_stats['python_percentage'] = (
    age_stats['python_users'] / age_stats['total_respondents'] * 100
)
age_stats
```
